#New model

For the new models we will be using optimised versions of the ckassification techniques we used before. For this purpose we will be using AdaBoost and XGBoost for further classification.

```python
import xgboost as xgb
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
```

#AdaBoost

```python
base_estimator = DecisionTreeClassifier(max_depth=7)

model = AdaBoostClassifier(base_estimator=base_estimator, n_estimators=200, learning_rate=0.1)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
report = classification_report(y_test, y_pred)
print(report)
```

#XGBoost

```python
dtrain = xgb.DMatrix(X_train, label=y_train)
dtest = xgb.DMatrix(X_test, label=y_test)

params = {
    'max_depth': 7,
    'eta': 0.1,
    'objective': 'multi:softmax',
    'num_class': 3
}

model = xgb.train(params, dtrain, num_boost_round=100)

y_pred = model.predict(dtest)
y_pred = [round(value) for value in y_pred]

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```

```python
report = classification_report(y_test, y_pred)
print(report)
```

#XGB without intermediate

```python
dtrain = xgb.DMatrix(X_train, label=y_train)
dtest = xgb.DMatrix(X_test, label=y_test)

params = {
    'max_depth': 5,
    'eta': 0.05,
    'objective': 'binary:logistic',
    'eval_metric': 'logloss'
}

model = xgb.train(params, dtrain, num_boost_round=100)

y_pred = model.predict(dtest)
y_pred = [round(value) for value in y_pred]

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")
```
